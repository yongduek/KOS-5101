```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# 1. 재현성을 위한 시드 설정 및 Generator 활용
rng = np.random.default_rng(seed=42)
loc, scale, size = 0, 1, 4000
samples = rng.normal(loc=loc, scale=scale, size=size)
```

```python
# 2. 시각화 설정
plt.figure(figsize=(9, 6))
count, bins, ignored = plt.hist(samples, bins=50, density=True, alpha=0.6, edgecolor='black', label='Sample Histogram')
```

```python
# 3. scipy를 활용한 이론적 PDF 계산 (범위 자동 설정)
x = np.linspace(samples.min(), samples.max(), 200)
pdf = norm.pdf(x, loc, scale)

plt.plot(x, pdf, 'r-', linewidth=2.5, label=f'Theoretical PDF (N({loc}, {scale}^2))')
```

```python
# 4. 부가 정보 추가
plt.title(f'Normal Distribution Analysis (n={size})', fontsize=14)
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

# 신뢰구간(Confidence Interval)과 신용구간(Credible Interval)의 개념과 비교

> **학습 목표**: 빈도주의적 통계의 신뢰구간과 베이지안 통계의 신용구간(크레디블 구간)이 철학적·수학적으로 어떻게 다른지를 이해한다.

---

두 가지 구간 추정 방법은 겉으로 유사해 보이지만, 확률에 대한 근본적으로 다른 해석에 기초하고 있다.

| 구분 | 신뢰구간 (Confidence Interval, CI) | 신용구간 (Credible Interval, CrI) |
|---|---|---|
| 통계 패러다임 | 빈도주의 (Frequentist) | 베이지안 (Bayesian) |
| 모수의 성격 | 고정된 미지의 상수 | 확률 변수 |
| 확률의 의미 | 반복 표집에서의 장기 비율 | 특정 사건에 대한 믿음의 정도 |
| 사전 정보 활용 | 불가 | 가능 (사전분포) |

## 1. 신뢰구간 (Confidence Interval)

### 1.1 통계적 기반: 빈도주의

신뢰구간은 **빈도주의(frequentist)** 패러다임에서 비롯된 개념이다. 빈도주의에서는 확률을 **무한히 반복된 실험에서 사건이 발생하는 장기적 비율(long-run relative frequency)**로 정의한다. 모수 $\theta$는 알 수 없지만 고정된 상수로 간주하며, 데이터와 추정량(estimator)만이 확률적 변동을 갖는다.

### 1.2 수학적 정의

확률표본 $X_1, X_2, \ldots, X_n \overset{iid}{\sim} f(x;\theta)$에 대해, 두 통계량 $L(\mathbf{X})$과 $U(\mathbf{X})$가 다음을 만족할 때 $[L, U]$를 $\theta$에 대한 $100(1-\alpha)\%$ **신뢰구간(CI)**이라 한다.

$$P\bigl(L(\mathbf{X}) \le \theta \le U(\mathbf{X})\bigr) = 1 - \alpha$$

여기서 $\alpha$는 유의수준(significance level)이며, $1 - \alpha$는 **신뢰수준(confidence level)**이다.

정규모집단 $X_i \sim \mathcal{N}(\mu, \sigma^2)$에서 모분산 $\sigma^2$이 알려진 경우, 모평균 $\mu$에 대한 $95\%$ 신뢰구간은 다음과 같다.

$$\hat{\mu}_{\text{CI}} = \bar{X} \pm z_{0.025} \cdot \frac{\sigma}{\sqrt{n}}, \qquad z_{0.025} = 1.96$$

### 1.3 올바른 해석

신뢰구간을 해석할 때 빈번히 발생하는 **오류**와 **정확한 해석**은 다음과 같다.

**잘못된 해석 (흔한 오해):**
> "95% 신뢰구간 $[2.3,\; 5.7]$은 모수 $\theta$가 이 구간에 속할 확률이 95%라는 것을 의미한다."

이는 틀린 해석이다. 빈도주의 틀에서 모수 $\theta$는 고정된 상수이므로, $\theta$에 대한 확률 진술 자체가 성립하지 않는다. $\theta$는 구간에 속하거나 속하지 않을 뿐이다.

**올바른 해석:**
> "동일한 모집단에서 반복 표집하여 같은 방법으로 구간을 구성할 경우, 장기적으로 그 구간들의 **$95\%$가 참 모수 $\theta$를 포함한다**."

즉, 확률 $1-\alpha$는 **구간 구성 절차(procedure)**의 성질이지, 특정 구간에 $\theta$가 있을 가능성이 아니다.

### 1.4 핵심 성질

- 모수 $\theta$의 사전 분포를 가정하지 않으므로 사전 정보(prior information)를 반영할 수 없다.
- 확률은 **데이터가 주어지기 전**, 표본 공간 위에서 정의된다.  
- 신뢰수준 $1-\alpha$를 높이면 구간의 폭이 넓어진다.  
- 표본 크기 $n$이 커질수록 구간의 폭은 $\mathcal{O}(n^{-1/2})$의 속도로 좁아진다.

## 2. 신용구간 (Credible Interval)

### 2.1 통계적 기반: 베이지안

신용구간(credible interval, CrI)은 **베이지안(Bayesian)** 패러다임에서 비롯된 개념이다. 베이지안에서는 확률을 **불확실성에 대한 믿음의 정도(degree of belief)**로 정의하며, 모수 $\theta$를 고정된 상수가 아닌 **확률 변수**로 취급한다. 따라서 "$\theta$가 특정 구간에 속할 확률"이라는 진술이 직접적으로 의미를 가진다.

### 2.2 베이즈 정리를 통한 사후분포 도출

분석의 출발점은 **사전분포(prior distribution)** $p(\theta)$이다. 이는 데이터를 관측하기 이전에 $\theta$에 대해 갖고 있는 믿음을 나타낸다. 데이터 $\mathbf{y} = (y_1, \ldots, y_n)$을 관측한 후, **베이즈 정리(Bayes' theorem)**를 적용하여 **사후분포(posterior distribution)** $p(\theta \mid \mathbf{y})$를 얻는다.

$$p(\theta \mid \mathbf{y}) = \frac{p(\mathbf{y} \mid \theta)\; p(\theta)}{p(\mathbf{y})} \propto \underbrace{p(\mathbf{y} \mid \theta)}_{\text{우도(likelihood)}} \times \underbrace{p(\theta)}_{\text{사전분포(prior)}}$$

여기서 $p(\mathbf{y}) = \int p(\mathbf{y} \mid \theta)\, p(\theta)\, d\theta$는 **주변 우도(marginal likelihood)** 또는 **증거(evidence)**로 불리며, $\theta$에 무관한 정규화 상수(normalizing constant)이다.

### 2.3 수학적 정의

사후분포 $p(\theta \mid \mathbf{y})$로부터, 두 값 $l$과 $u$가 다음을 만족할 때 $[l, u]$를 $\theta$에 대한 $100(1-\alpha)\%$ **신용구간(CrI)**이라 한다.

$$P\bigl(l \le \theta \le u \mid \mathbf{y}\bigr) = \int_{l}^{u} p(\theta \mid \mathbf{y})\, d\theta = 1 - \alpha$$

이 정의에서 데이터 $\mathbf{y}$는 이미 **관측된 고정값**이며, 확률 진술은 **$\theta$에 관한 것**임을 주목하라. 이 점이 신뢰구간과 근본적으로 다른 부분이다.

### 2.4 대표적인 신용구간 유형

신용구간을 구성하는 방법은 유일하지 않다. 대표적인 두 가지 방법은 다음과 같다.

**(1) 등미확구간 (Equal-tailed Credible Interval, ETI)**

사후분포의 양쪽 꼬리에 각각 $\alpha/2$의 확률을 배분하는 방식이다.

$$l = Q_{\alpha/2}(\theta \mid \mathbf{y}), \qquad u = Q_{1-\alpha/2}(\theta \mid \mathbf{y})$$

여기서 $Q_p$는 사후분포의 $p$번째 분위수(quantile)를 의미한다. 구성이 간단하고 중앙값을 중심으로 대칭적인 구간을 형성한다.

**(2) 최고 사후 밀도 구간 (Highest Posterior Density Interval, HPDI)**

동일한 신뢰 수준 $1-\alpha$ 하에서 **구간의 폭을 최소화**하는 방식이다. 구간 내 모든 점의 사후 밀도가 구간 외부의 모든 점보다 크거나 같도록 구성한다.

$$[l^*, u^*] = \arg\min_{[l,u]: P(l \le \theta \le u \mid \mathbf{y}) = 1-\alpha} (u - l)$$

사후분포가 비대칭이거나 다봉(multimodal)일 때, HPDI는 ETI보다 더 정보량이 높은 구간을 제공한다.

### 2.5 올바른 해석

**올바른 해석:**
> "관측된 데이터 $\mathbf{y}$가 주어졌을 때, 모수 $\theta$가 구간 $[l, u]$에 속할 **사후 확률이 $95\%$이다**."

이 진술은 직관적이며 연구자들이 실제로 원하는 해석이기도 하다. 신용구간에서는 구간 $[l, u]$가 데이터에 의해 고정되어 있고, $\theta$의 불확실성이 사후분포를 통해 확률로 표현된다.

### 2.6 핵심 성질

- **사전분포의 선택이 결과에 영향**을 미친다. 사전 정보가 충분하고 합리적일수록 추론이 개선된다.  
- 표본 크기 $n$이 충분히 크면, 합리적인 범위 내에서 사전분포의 영향이 줄어들고 **사후분포는 우도(likelihood)에 의해 지배**된다 (Bernstein–von Mises 정리).  
- 확률은 **데이터가 주어진 후**, 모수 공간 위에서 정의된다.  
- 폐쇄형 해(closed-form solution)가 없는 경우 **MCMC(Markov Chain Monte Carlo)**와 같은 수치적 방법으로 사후분포를 근사한다.

## 3. 두 구간의 비교 및 철학적 차이

### 3.1 핵심 차이 요약

| 비교 항목 | 신뢰구간 (CI) | 신용구간 (CrI) |
|---|---|---|
| 통계 패러다임 | 빈도주의 | 베이지안 |
| 모수 $\theta$의 성격 | 고정된 미지의 상수 | 확률 변수 |
| 확률이 정의되는 대상 | 구간(반복 표집의 성질) | 모수 $\theta$ (믿음의 정도) |
| 사전 정보 반영 | 불가 | 사전분포 $p(\theta)$로 반영 |
| 해석 | 절차의 장기 성질 | 데이터 조건부 확률 진술 |
| 직관적 해석 가능성 | 비직관적 | 직관적 |
| 수치적 도구 | 닫힌 형식(closed-form) | MCMC, 변분 추론(VI) 등 |

### 3.2 동일한 표현, 다른 의미

두 구간 모두 "$95\%$ 구간"이라는 동일한 표현을 사용하지만 그 의미는 전혀 다르다.

- **신뢰구간**: $P(L(\mathbf{X}) \le \theta \le U(\mathbf{X})) = 0.95$  
  → 확률은 **데이터$(\mathbf{X})$**에 걸려 있다. 수집되기 전의 표본이 확률적이기 때문이다.

- **신용구간**: $P(l \le \theta \le u \mid \mathbf{y}) = 0.95$  
  → 확률은 **모수$(\theta)$**에 걸려 있다. 데이터 $\mathbf{y}$는 이미 관측된 고정값이다.

### 3.3 언제 어느 구간을 사용하는가?

**신뢰구간이 적합한 경우:**
- 사전 정보가 없거나, 분석에 주관적 판단을 개입시키고 싶지 않을 때
- 빈도주의적 가설 검정(NHST)과 일관된 틀을 유지하고자 할 때
- 표본 수가 많아 점근적 정규 근사가 타당할 때

**신용구간이 적합한 경우:**
- 모수에 대한 직접적인 확률 진술이 필요할 때 (예: "효과 크기가 양수일 확률")
- 선행 연구, 전문가 지식 등 사전 정보를 분석에 통합하고자 할 때
- 표본 수가 적어 점근적 근사가 부적절할 때
- 복잡한 계층 모형(hierarchical model)이나 결측 데이터 등 복잡한 구조적 문제를 다룰 때

### 3.4 대표적 오개념 정리

> **오개념**: "95% 신뢰구간이 계산되었으므로, 참 모수가 이 구간 안에 있을 확률은 95%이다."

이는 신뢰구간에 대한 가장 흔하고 심각한 오해이다. 특정 표본에서 계산된 구간 $[l, u]$에 대해 $\theta$는 그 구간 안에 있거나 그렇지 않거나 둘 중 하나이다. "95%"라는 수치는 이 특정 구간에 적용되는 것이 아니라, **구간을 생성하는 방법(procedure)**의 장기적 성질이다.

이 진술이 성립하려면 $\theta$가 확률 변수이어야 하며, 이는 베이지안의 전제조건이다. 따라서 모수에 대한 직접적인 확률 진술은 **신용구간**을 통해서만 올바르게 표현될 수 있다.

## 4. 수치 예제 및 시뮬레이션

### 설정 (Generative Model)

다음과 같은 정규 모형을 가정한다.

$$X_i \sim \mathcal{N}(\mu,\; \sigma^2), \quad i = 1, \ldots, n$$

- 참 모수: $\mu = 5$ (분석자에게 미지)
- 모분산(既知): $\sigma^2 = 4$ ($\sigma = 2$)
- 표본 크기: $n = 20$

이 설정에서 $95\%$ 신뢰구간의 반폭(half-width)은 다음과 같다.

$$z_{0.025} \cdot \frac{\sigma}{\sqrt{n}} = 1.96 \times \frac{2}{\sqrt{20}} \approx 0.876$$

---

### 4.1 신뢰구간 — 반복 표집 시뮬레이션

"95% 신뢰구간"의 올바른 의미는 반복 표집(repeated sampling)을 통해 가장 직관적으로 확인할 수 있다.  
$B = 200$회 반복하여 각 표본에서 95% CI를 계산하고, 참 모수 $\mu = 5$를 포함하는 비율을 확인한다.

---

### 4.2 베이지안 신용구간 — 켤레 사전분포 분석

모분산 $\sigma^2$이 알려진 정규 모형에서 모평균 $\mu$에 대한 **켤레 사전분포(conjugate prior)**는 정규분포이다.

$$\mu \sim \mathcal{N}(\mu_0,\; \tau_0^2)$$

사전분포와 우도를 결합하면 사후분포(posterior) 역시 정규분포이다.

$$\mu \mid \mathbf{y} \sim \mathcal{N}(\mu_n,\; \tau_n^2)$$

여기서 사후 모수(posterior parameters)는 다음과 같다.

$$\frac{1}{\tau_n^2} = \frac{1}{\tau_0^2} + \frac{n}{\sigma^2}, \qquad \mu_n = \tau_n^2 \left(\frac{\mu_0}{\tau_0^2} + \frac{n\bar{y}}{\sigma^2}\right)$$

직관적으로, 사후 평균 $\mu_n$은 **사전 평균 $\mu_0$와 표본 평균 $\bar{y}$의 정밀도(precision) 가중 평균**이다.  
표본 크기 $n$이 커질수록 사후 평균은 표본 평균에 수렴하고, 사前 정보의 영향은 감소한다.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import stats

# ── 공통 설정 ──────────────────────────────────────────────
rng   = np.random.default_rng(seed=42)
mu    = 5.0      # 참 모수 (true parameter)
sigma = 2.0      # 알려진 모표준편차 (known population SD)
n     = 20       # 표본 크기
alpha = 0.05     # 유의수준
z     = stats.norm.ppf(1 - alpha / 2)   # z_{0.025} ≈ 1.96

print(f"참 모수 μ = {mu}")
print(f"σ = {sigma}, n = {n}, z_{{0.025}} = {z:.4f}")
print(f"95% CI 반폭 = {z * sigma / np.sqrt(n):.4f}")
```

```python
# -- 4.1 Confidence interval repeated-sampling simulation ----------------
B = 200  # number of repetitions
half_width = z * sigma / np.sqrt(n)

# Repeat B times:
# 1) sample n observations from the data-generating distribution
# 2) compute sample mean
# 3) compute the 95% CI from that sample
xbars = np.empty(B)
lowers = np.empty(B)
uppers = np.empty(B)

for i in range(B):
    sample_i = rng.normal(loc=mu, scale=sigma, size=n)
    xbar_i = sample_i.mean()
    xbars[i] = xbar_i
    lowers[i] = xbar_i - half_width
    uppers[i] = xbar_i + half_width

contains_mu = (lowers <= mu) & (mu <= uppers)
coverage = contains_mu.mean()

# -- Visualization -------------------------------------------------------
fig, ax = plt.subplots(figsize=(5, 9))

for i in range(B):
    color = "#2196F3" if contains_mu[i] else "#F44336"
    lw = 0.8
    ax.plot([lowers[i], uppers[i]], [i, i], color=color, lw=lw, solid_capstyle="round")
    ax.plot(xbars[i], i, "o", color=color, markersize=4)

ax.axvline(mu, color="black", lw=1.5, linestyle="--", label=f"True $\\mu = {mu}$")

patch_in = mpatches.Patch(color="#2196F3", label=f"Contains $\\mu$ ({contains_mu.sum()})")
patch_out = mpatches.Patch(color="#F44336", label=f"Misses $\\mu$ ({(~contains_mu).sum()})")
ax.legend(handles=[patch_in, patch_out], loc="lower right", fontsize=9)

ax.set_xlabel("Parameter space", fontsize=11)
ax.set_ylabel(f"Sample index (total {B})", fontsize=11)
ax.set_title(
    f"95% Confidence Interval: Repeated Sampling Simulation\n"
    f"Actual coverage = {coverage:.1%}  (theoretical ~95%)",
    fontsize=12,
)
ax.set_yticks([])
plt.tight_layout()
plt.show()

print(f"\n[Result] Out of {B} intervals, {contains_mu.sum()} contain true μ={mu} ({coverage:.1%}).")
```

```python
# ── 4.2 베이지안 신용구간 — 켤레 사전분포 분석 ─────────────
# 하나의 관측 표본을 사용
rng_fixed = np.random.default_rng(seed=7)
y_obs     = rng_fixed.normal(loc=mu, scale=sigma, size=n)
y_bar     = y_obs.mean()

# 사전분포 설정 (두 가지 비교)
priors = [
    {"label": "Informative prior\n$\\mu_0=3,\\;\\tau_0=1$", "mu0": 3.0, "tau0": 1.0, "color": "#9C27B0"},
    {"label": "Diffuse prior\n$\\mu_0=0,\\;\\tau_0=10$", "mu0": 0.0, "tau0": 10.0, "color": "#FF9800"},
]

theta_grid = np.linspace(0, 10, 1000)

fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=False)

for ax, pr in zip(axes, priors):
    mu0, tau0 = pr["mu0"], pr["tau0"]

    # 사후 정밀도 및 사후 평균 (켤레 공식)
    prec_prior = 1 / tau0**2
    prec_like  = n / sigma**2
    tau_n2     = 1 / (prec_prior + prec_like)
    mu_n       = tau_n2 * (mu0 * prec_prior + y_bar * prec_like)
    tau_n      = np.sqrt(tau_n2)

    # 95% ETI (등미확 신용구간)
    cri_lo = stats.norm.ppf(alpha / 2,     loc=mu_n, scale=tau_n)
    cri_hi = stats.norm.ppf(1 - alpha / 2, loc=mu_n, scale=tau_n)

    # 95% 신뢰구간 (같은 데이터)
    ci_lo = y_bar - z * sigma / np.sqrt(n)
    ci_hi = y_bar + z * sigma / np.sqrt(n)

    # 사전·사후·우도 밀도
    prior_pdf     = stats.norm.pdf(theta_grid, mu0, tau0)
    posterior_pdf = stats.norm.pdf(theta_grid, mu_n, tau_n)

    ax.plot(theta_grid, prior_pdf,     "--", color=pr["color"], lw=1.5, label="Prior $p(\\mu)$")
    ax.plot(theta_grid, posterior_pdf, "-",  color=pr["color"], lw=2.5, label="Posterior $p(\\mu|\\mathbf{y})$")

    # 95% 신용구간 음영
    mask = (theta_grid >= cri_lo) & (theta_grid <= cri_hi)
    ax.fill_between(theta_grid, posterior_pdf, where=mask,
                    alpha=0.25, color=pr["color"], label=f"95% CrI [{cri_lo:.2f}, {cri_hi:.2f}]")

    # 참 모수 및 표본 평균 표시
    ax.axvline(mu,    color="black",   lw=1.5, linestyle="--", label=f"True $\\mu={mu}$")
    ax.axvline(y_bar, color="#4CAF50", lw=1.5, linestyle=":" , label=f"Sample mean $\\bar{{y}}={y_bar:.2f}$")

    # 95% CI 구간 표시 (x-축 아래)
    ax.annotate("", xy=(ci_hi, -0.04), xytext=(ci_lo, -0.04),
                xycoords=("data", "axes fraction"), textcoords=("data", "axes fraction"),
                arrowprops=dict(arrowstyle="<->", color="gray", lw=1.5))
    ax.text((ci_lo + ci_hi) / 2, -0.09, f"95% CI\n[{ci_lo:.2f}, {ci_hi:.2f}]",
            ha="center", va="top", fontsize=8, color="gray",
            transform=ax.get_xaxis_transform())

    ax.set_title(pr["label"], fontsize=12)
    ax.set_xlabel("$\\mu$", fontsize=11)
    ax.set_ylabel("Density", fontsize=11)
    ax.legend(fontsize=8, loc="upper right")
    ax.set_xlim(0, 10)

    print(f"[{pr['label'].replace(chr(10),' ')}]")
    print(f"  사후 평균 μₙ = {mu_n:.4f},  사후 SD τₙ = {tau_n:.4f}")
    print(f"  95% CrI  : [{cri_lo:.4f}, {cri_hi:.4f}]")
    print(f"  95% CI   : [{ci_lo:.4f},  {ci_hi:.4f}]")
    print()

fig.suptitle(
    f"Bayesian Credible Interval (CrI)  vs  Confidence Interval (CI)\n"
    f"Data: $n={n}$, $\\bar{{y}}={y_bar:.2f}$, $\\sigma={sigma}$ (known)",
    fontsize=13,
)
plt.tight_layout()
plt.show()
```

```python
# ── 4.3 표본 크기 증가에 따른 사전분포 영향 소멸 ─────────────
# Bernstein–von Mises 정리 시각화:
# n이 충분히 크면 모든 합리적 사전분포에서 사후분포 → 우도

ns       = [5, 20, 100, 500]
mu0_inf  = 3.0   # 정보성 사전 평균 (참 μ=5 와 의도적으로 다르게 설정)
tau0_inf = 1.0
theta_g  = np.linspace(1, 9, 1000)

fig, axes = plt.subplots(1, len(ns), figsize=(14, 3.5), sharey=False)

for ax, ni in zip(axes, ns):
    obs_i  = rng_fixed.normal(loc=mu, scale=sigma, size=ni)
    ybar_i = obs_i.mean()

    # 사후 모수
    prec_n  = 1 / tau0_inf**2 + ni / sigma**2
    tau_n_i = 1 / np.sqrt(prec_n)
    mu_n_i  = (mu0_inf / tau0_inf**2 + ni * ybar_i / sigma**2) / prec_n

    # 95% CrI 와 CI
    cri_lo_i = stats.norm.ppf(0.025, mu_n_i, tau_n_i)
    cri_hi_i = stats.norm.ppf(0.975, mu_n_i, tau_n_i)
    ci_lo_i  = ybar_i - z * sigma / np.sqrt(ni)
    ci_hi_i  = ybar_i + z * sigma / np.sqrt(ni)

    post_pdf = stats.norm.pdf(theta_g, mu_n_i, tau_n_i)
    ax.plot(theta_g, post_pdf, lw=2, color="#9C27B0", label="Posterior")

    mask = (theta_g >= cri_lo_i) & (theta_g <= cri_hi_i)
    ax.fill_between(theta_g, post_pdf, where=mask, alpha=0.3, color="#9C27B0", label=f"95% CrI")

    ax.axvline(mu,     lw=1.5, linestyle="--", color="black",   label=f"True μ={mu}")
    ax.axvline(ybar_i, lw=1.5, linestyle=":" , color="#4CAF50", label=f"Sample mean ȳ={ybar_i:.2f}")

    ax.set_title(f"$n = {ni}$\nCrI=[{cri_lo_i:.2f},{cri_hi_i:.2f}]\nCI=[{ci_lo_i:.2f},{ci_hi_i:.2f}]",
                 fontsize=9)
    ax.set_xlabel("$\\mu$", fontsize=10)
    ax.set_yticks([])
    if ni == ns[0]:
        ax.set_ylabel("Density", fontsize=10)
    ax.set_xlim(1, 9)

axes[0].legend(fontsize=7, loc="upper right")
fig.suptitle(
    f"Effect of increasing $n$: Prior influence diminishes (Bernstein\u2013von Mises)\n"
    f"Informative prior: $\\mu_0={mu0_inf},\\;\\tau_0={tau0_inf}$ (offset from true $\\mu={mu}$)",
    fontsize=12,
)
plt.tight_layout()
plt.show()
```

### 시뮬레이션 결과 해석

**4.1 반복 표집 시뮬레이션 (신뢰구간)**

- $B = 200$번의 반복 중 약 $95\%$의 구간이 참 모수 $\mu = 5$를 포함한다.  
- 파란색 구간은 $\mu$를 포함하고, 빨간색 구간은 포함하지 않는다.  
- 특정 구간이 $\mu$를 포함하는지는 해당 구간을 얻고 나서야 알 수 있으며, 그 시점에서는 포함 여부가 0 또는 1이다 (확률이 아님).  
- 이것이 "95%"가 **구간에 적용되지 않고 절차에 적용된다**는 의미이다.

**4.2 켤레 사전분포를 이용한 베이지안 신용구간**

- **정보 사전분포** ($\mu_0 = 3,\;\tau_0 = 1$): 사전 평균이 참 값과 다를 때, 사후 평균은 사전 평균 쪽으로 편향된다.  
  사후분포의 $95\%$ 신용구간은 "이 구간에 $\mu$가 속할 사후 확률이 $95\%$"라는 직접적인 의미를 갖는다.
- **산만 사전분포** ($\mu_0 = 0,\;\tau_0 = 10$): 사전 정보가 거의 없으므로 사후분포가 우도에 의해 지배된다.  
  결과적으로 신용구간이 신뢰구간과 매우 유사해진다.

**4.3 표본 크기 증가**

- 표본 크기 $n$이 증가할수록, 사전분포가 정보성이더라도 사후분포는 점차 우도(likelihood) 중심으로 집중된다.  
- 이는 **Bernstein–von Mises 정리**의 직관적 확인이다:  
  $n \to \infty$이면 $p(\mu \mid \mathbf{y}) \approx \mathcal{N}(\hat{\mu}_{\text{MLE}},\; I_n(\mu)^{-1})$  
  여기서 $I_n(\mu)$는 피셔 정보량(Fisher information)이다.
- 따라서 충분히 큰 표본에서는 신용구간과 신뢰구간이 수치적으로 거의 일치하지만, **해석은 여전히 근본적으로 다르다.**

---

> **핵심 교훈**: 두 구간이 동일한 수치를 산출하든 아니든, 그 수치가 의미하는 바는 근본적으로 다르다.  
> 신뢰구간의 "95%"는 **반복 표집 절차**의 장기적 성질이다.  
> 신용구간의 "95%"는 **지금 주어진 데이터 하에서 모수가 그 구간에 있을 확률**이다.

## 5. Confidence Interval with Unknown $\sigma$: t-interval

모표준편차 $\sigma$를 모르는 상황에서, 표본표준편차를 사용해 각 반복마다 t-구간을 구성하는 절차를 시뮬레이션한다.

When population standard deviation $\sigma$ is unknown, each repeated sample must estimate variability using the sample standard deviation $s_i$.

For each repetition $i=1,\dots,B$:

$$\bar{X}_i = \frac{1}{n}\sum_{j=1}^n X_{ij}, \qquad s_i = \sqrt{\frac{1}{n-1}\sum_{j=1}^n (X_{ij}-\bar{X}_i)^2}$$

$$\mathrm{CI}_i^{(t)} = \bar{X}_i \pm t_{1-\alpha/2,\;n-1}\cdot \frac{s_i}{\sqrt{n}}$$

This section uses the same repeated-sampling structure as Section 4.1, but replaces the z-based half-width with sample-specific t-based half-width.

```python
# -- Section 5: t-interval repeated-sampling simulation (sigma unknown) --
import scipy.stats as stats

B_t = 200
n_t = n
alpha = 0.05 # 유의수준, 1-alpha = 0.95
t_crit = stats.t.ppf(1 - alpha/2, df=n_t - 1) 

xbars_t = np.empty(B_t)
svals_t = np.empty(B_t)
lowers_t = np.empty(B_t)
uppers_t = np.empty(B_t)

for i in range(B_t):
    sample_i = rng.normal(loc=mu, scale=sigma, size=n_t)
    xbar_i = sample_i.mean()
    s_i = sample_i.std(ddof=1) # ddof=1: 표본 표준편차 계산에서 n-1로 나누기
    half_width_i = t_crit * s_i / np.sqrt(n_t)

    xbars_t[i] = xbar_i
    svals_t[i] = s_i
    lowers_t[i] = xbar_i - half_width_i
    uppers_t[i] = xbar_i + half_width_i

contains_mu_t = (lowers_t <= mu) & (mu <= uppers_t)
coverage_t = contains_mu_t.mean()

fig, ax = plt.subplots(figsize=(5, 9))

for i in range(B_t):
    color = "#1976D2" if contains_mu_t[i] else "#E53935"
    ax.plot([lowers_t[i], uppers_t[i]], [i, i], color=color, lw=0.8, solid_capstyle="round")

ax.axvline(mu, color="black", lw=1.5, linestyle="--", label=f"True $\\mu = {mu}$")

patch_in_t = mpatches.Patch(color="#1976D2", label=f"Contains $\\mu$ ({contains_mu_t.sum()})")
patch_out_t = mpatches.Patch(color="#E53935", label=f"Misses $\\mu$ ({(~contains_mu_t).sum()})")
ax.legend(handles=[patch_in_t, patch_out_t], loc="lower right", fontsize=9)

ax.set_xlabel("Parameter space", fontsize=11)
ax.set_ylabel(f"Sample index (total {B_t})", fontsize=11)
ax.set_title(
    f"95% t-interval: Repeated Sampling Simulation (sigma unknown)\n"
    f"Actual coverage = {coverage_t:.1%}  (target ~95%, df={n_t-1})",
    fontsize=12,
)
ax.set_yticks([])
plt.tight_layout()
plt.show()

print(f"[t-interval] Out of {B_t} intervals, {contains_mu_t.sum()} contain true μ={mu} ({coverage_t:.1%}).")
print(f"Average sample SD across repetitions: {svals_t.mean():.4f}")
```

## 6. Bayesian Posterior for Unknown $\sigma$ via Stan/cmdstanpy

모표준편차 $\sigma$를 모르는 상황에서, Stan 모델을 사용해 $(\mu,\sigma)$의 사후분포를 표집하고 각 파라메터의 $96\%$ 신용구간(credible interval)을 계산한다.

본 셀은 다음 순서로 진행한다.

1. 관측 데이터 $\mathbf{y}$를 구성한다.
2. Stan 모델을 컴파일하고 NUTS로 사후표본을 추출한다.
3. $\mu$와 $\sigma$ 각각에 대해 분위수 기반 $96\%$ CrI ($2\%$~$98\%$)를 계산한다.
4. 사후분포 히스토그램 위에 CrI 경계와 중앙값을 시각화한다.

```python
# -- Section 6: Stan/cmdstanpy posterior sampling for (mu, sigma) ---------
from pathlib import Path
import tempfile

try:
    from cmdstanpy import CmdStanModel, cmdstan_path
except ImportError as exc:
    raise ImportError(
        "cmdstanpy is not installed. Install it first, e.g. `pip install cmdstanpy`."
    ) from exc


def hpdi(samples, cred_mass=0.96):
    """Return shortest interval [low, high] containing cred_mass of posterior samples."""
    s = np.sort(np.asarray(samples))
    n_s = len(s)
    m = int(np.floor(cred_mass * n_s))
    if m < 1:
        raise ValueError("Too few samples for HPDI computation.")
    widths = s[m:] - s[: n_s - m]
    j = np.argmin(widths)
    return s[j], s[j + m]


# Check CmdStan toolchain availability.
_ = cmdstan_path()

# Observed data for unknown-sigma case (consistent with prior sections)
rng_stan = np.random.default_rng(20260409) # random number generator for reproducibility
y_obs_stan = rng_stan.normal(loc=mu, scale=sigma, size=n) # observed data (n=20)
N_stan = len(y_obs_stan)

stan_code = r"""
data {
  int<lower=1> N;
  array[N] real y;
}
parameters {
  real mu;
  real<lower=0> sigma;
}
model {
  mu ~ normal(0, 10); 
  sigma ~ exponential(1);
  y ~ normal(mu, sigma);
}
"""

with tempfile.TemporaryDirectory() as tmpdir:
    stan_file = Path(tmpdir) / "normal_unknown_sigma.stan"
    stan_file.write_text(stan_code, encoding="utf-8")

    model = CmdStanModel(stan_file=str(stan_file))
    fit = model.sample(
        data={"N": N_stan, "y": y_obs_stan.tolist()},
        chains=4,
        parallel_chains=4,
        iter_warmup=1000,
        iter_sampling=2000,  # 더 많은 샘플링 반복으로 HPDI 안정화; Just for demonstration, not too large to avoid long runtime
        seed=20260409,
        show_progress=False,
    )

posterior = fit.draws_pd()[["mu", "sigma"]]
mu_draws = posterior["mu"].to_numpy()
sigma_draws = posterior["sigma"].to_numpy()

# 96% HPDI only
mu_hpdi_low, mu_hpdi_high = hpdi(mu_draws, cred_mass=0.96)
sig_hpdi_low, sig_hpdi_high = hpdi(sigma_draws, cred_mass=0.96)

# Posterior summaries for arrow markers
mu_mean = mu_draws.mean()
mu_median = np.median(mu_draws)
sig_mean = sigma_draws.mean()
sig_median = np.median(sigma_draws)

print("96% HPDI from Stan posterior")
print(f"mu    HPDI: [{mu_hpdi_low:.4f}, {mu_hpdi_high:.4f}]  (mean = {mu_mean:.4f}, median = {mu_median:.4f})")
print(f"sigma HPDI: [{sig_hpdi_low:.4f}, {sig_hpdi_high:.4f}]  (mean = {sig_mean:.4f}, median = {sig_median:.4f})")
```

### + 왜 $\sigma$의 사전분포로 $\mathrm{Exponential}(1)$를 자주 쓰는가?

$\sigma$는 표준편차이므로 반드시 양수여야 한다. 따라서 지지집합이 $(0,\infty)$인 지수분포는 형식적으로 자연스럽다.

$\sigma \sim \mathrm{Exponential}(1)$를 기본값으로 자주 권장하는 이유는 다음과 같다.

1. **양의 제약을 정확히 반영**  
   $\sigma>0$ 조건을 사전분포 자체가 자동으로 보장한다.

2. **약한 정규화(weakly informative regularization)**  
   작은 표본에서 비현실적으로 큰 분산으로 추정이 튀는 현상을 완화한다. 즉, 추론을 안정화하는 방향으로 작동한다.

3. **큰 값도 완전히 배제하지 않음**  
   꼬리가 있어 데이터가 강하게 요구하면 큰 $\sigma$도 허용한다. 따라서 과도하게 경직된 사전분포는 아니다.

4. **HMC/NUTS 계산 안정성에 유리한 편**  
   매우 무거운 꼬리(heavy tail) 사전분포보다 샘플링 기하가 안정적인 경우가 많아, 실제 계산에서 수렴 및 혼합이 개선되는 경향이 있다.

5. **해석과 스케일 조정이 단순함**  
   일반적으로 $\sigma \sim \mathrm{Exponential}(\lambda)$에서 평균은 $E[\sigma]=1/\lambda$이다.  
   따라서 데이터 스케일이 1이 아니면 $\lambda$를 조정해 같은 형태로 쉽게 보정할 수 있다.

---

요약하면, $\mathrm{Exponential}(1)$은 “양수 제약 + 약한 정규화 + 계산 안정성”을 동시에 제공하는 실용적 기본값이다.  
다만 데이터의 단위/스케일이 크게 다르면 $\mathrm{Exponential}(1/s_0)$처럼 스케일 맞춤형으로 조정하는 것이 바람직하다.

### 왜 MAP가 참값(true value)과 멀어 보일 수 있는가?

현재 그림에서 MAP(최빈 사후점)가 참값 $(\mu=5,\;\sigma=2)$과 다르게 보이는 것은 **이상한 현상이라기보다 자연스러운 통계적 결과**이다. 핵심 이유는 다음과 같다.

1. **유한 표본 변동 (sampling variability)**  
   데이터는 $n=20$으로 크지 않다. 표본이 한 번 생성되면 우도의 꼭대기 자체가 참값에서 벗어날 수 있으며, MAP는 그 꼭대기(사전 포함 후)를 따르므로 참값과 차이가 난다.

2. **MAP는 "가장 높은 한 점"이지 중심이 아님**  
   MAP는 사후분포 전체의 평균이나 중앙값이 아니라, 밀도가 최대인 단일 점이다.  
   특히 2차원 $(\mu,\sigma)$에서는 분포의 비대칭/기울기 때문에 MAP가 posterior mean, median, true value와 서로 다른 위치에 올 수 있다.

3. **사전분포의 영향 (regularization)**  
   이 모델은
   $$\mu \sim \mathcal{N}(0,10),\qquad \sigma \sim \mathrm{Exponential}(1)$$
   를 사용한다. 특히 $\sigma$에 대한 지수 사전분포는 큰 값을 억제하고 작은 값 쪽에 더 많은 질량을 둔다. 데이터 정보가 강하지 않을 때 (데이터 갯수가 작을 때) MAP는 이 사전의 영향을 받아 이동할 수 있다.

4. **MAP는 매개변수화(parameterization)에 민감**  
   MAP는 변수 변환에 대해 불변이 아니다. 예를 들어 $\sigma$ 대신 $\log\sigma$를 중심으로 문제를 쓰면 최빈점 위치가 달라질 수 있다. 따라서 MAP만으로 "참값과의 거리"를 판단하면 과해석이 생길 수 있다.

5. **진단 기준은 한 점이 아니라 구간/영역 포함 여부**  
   Bayesian 추론에서는 단일 점보다 **신용구간(1D)**, **HDR 영역(2D)**이 더 중요하다.  
   현재 그림에서 참값이 50/80/90% HDR 내부에 있는지, 그리고 주변분포 96% CI에 포함되는지를 보는 것이 더 적절한 해석이다.

---

요약하면, MAP가 참값과 떨어져 보이는 것은 주로 **작은 표본에서의 우도 변동 + 사전 영향 + MAP의 점추정 특성** 때문에 발생한다.  
표본 크기 $n$이 커지면 일반적으로 posterior가 참값 주변으로 더 집중되고, MAP/mean/median의 차이도 줄어드는 경향이 있다.

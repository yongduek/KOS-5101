```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- 파라미터 설정 ---
mean = 5
std = 3


# 2. PDF 라인 계산: 이론적 곡선 도출
# x축 범위를 평균 주변 ±4표준편차로 설정 (데이터의 99.99% 포함 범위)
x = np.linspace(mean - 4*std, mean + 4*std, 1000)
pdf = norm.pdf(x, mean, std) # scipy.stats.norm의 pdf 메서드 사용

# 3. Pyplot 그리기 및 파일 저장
plt.figure(figsize=(8, 6))

# 히스토그램 (실제 데이터 분포)
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Samples')

# PDF 라인 (이론적 곡선)
plt.plot(x, pdf, 'r-', lw=2, label=f'Normal(mean={mean}, std={std})')

# 스타일링
plt.title('Normal Distribution Analysis')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, alpha=0.3)
```

```python
import scipy.stats as stats

mean = 5
std = 3
s1 = stats.norm.rvs(loc=mean, scale=std, size=1000)
s2 = stats.norm.rvs(loc=mean, scale=std, size=1000)
```

```python
plt.scatter(s1, s2, alpha=0.5)
plt.title('Scatter Plot of Samples')
```

```python
mean_s1 = s1.mean()
mean_s2 = s2.mean()

plt.scatter(mean_s1, mean_s2, color='red', marker='X', s=100, label='Mean Point')
plt.scatter(s1, s2, alpha=0.5, label='Samples')
plt.title('Scatter Plot of Samples with Mean Point')
plt.xlabel('s1')
plt.ylabel('s2')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
